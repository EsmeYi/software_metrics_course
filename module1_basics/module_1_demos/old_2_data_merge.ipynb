# Merging data from all trials to one large table

In this notebook we take all results from all trials (different network configurations) and we use that to make a statistics. 

```python
import pandas as pd
```

```python
dataset_1 = pd.read_csv("E:/ccflex_gerrit_experiments_axis/workspaces_repo/run/results_trial_1_dense_50_20_8/classified.csv", 
                        delimiter="$", low_memory=False)
dataset_2 = pd.read_csv("E:/ccflex_gerrit_experiments_axis/workspaces_repo/run/results_trial_2_dense_500_100_50/classified.csv", 
                        delimiter="$", low_memory=False)
dataset_3 = pd.read_csv("E:/ccflex_gerrit_experiments_axis/workspaces_repo/run/results_trial_3_dense_n_inputs_times_three/classified.csv", 
                        delimiter="$", low_memory=False)
dataset_4 = pd.read_csv("E:/ccflex_gerrit_experiments_axis/workspaces_repo/run/results/classified_cNN_trial_4.csv", 
                        delimiter="$", low_memory=False)
dataset_5 = pd.read_csv("E:/ccflex_gerrit_experiments_axis/workspaces_repo/run/results/classified_cNN_trial5.csv", 
                        delimiter="$", low_memory=False)
dataset_6 = pd.read_csv("E:/ccflex_gerrit_experiments_axis/workspaces_repo/run/results/classified_cNN_trial6.csv", 
                        delimiter="$", low_memory=False)
dataset_7 = pd.read_csv("E:/ccflex_gerrit_experiments_axis/workspaces_repo/run/results/classified_cNN_trial7.csv", 
                        delimiter="$", low_memory=False)
dataset_8 = pd.read_csv("E:/ccflex_gerrit_experiments_axis/classified_c50.csv", 
                        delimiter="$", low_memory=False)
dataset_9 = pd.read_csv("E:/ccflex_gerrit_experiments_axis/classified_cart.csv", 
                        delimiter="$", low_memory=False)
```

```python
all_data_sets = pd.concat([dataset_1, 
                           dataset_2[["Probability of violation"]], 
                           dataset_3[["Probability of violation"]], 
                           dataset_4[["Probability of violation"]], 
                           dataset_5[["Probability of violation"]], 
                           dataset_6[["Probability of violation"]], 
                           dataset_7[["Probability of violation"]],
                           dataset_8,
                           dataset_9], 
                           axis = 1,             
                          ignore_index = True)
all_data_sets.columns = ["index", 
                         "Filename", 
                         "Line", 
                         "Predicted violation", 
                         "Probability of violation (t1)", 
                         "Probability of violation (t2)", 
                         "Probability of violation (t3)", 
                         "Probability of violation (t4)", 
                         "Probability of violation (t5)", 
                         "Probability of violation (t6)", 
                         "Probability of violation (t7)", 
                         "Class (C50)", 
                         "Class (CART)"]
```
